# Plots of potential energy

## Here is the potential energy of a spring:

```python
k=20  # N/m
m=0.5 # kg
g=9.8 # N/kg
L=0.1 # m, relaxed spring length

x=[]
u_sp=[]
u_g=[]
u_total=[]

for i in range(30):
    position = L+(3*L)*(i/30)
    x.append(position)
    spring_energy=0.5*k*position**2
    gravitational_energy=-m*g*position
    u_sp.append(spring_energy)
    u_g.append(gravitational_energy)
    u_total.append(spring_energy+gravitational_energy)

import matplotlib.pyplot as plt
plt.plot(x,u_sp,label="$U_{Spring}$")
plt.plot(x,u_g,label="$U_{Gravity}$")
plt.plot(x,u_total,label="Total potential energy")
plt.legend()
plt.ylabel("Energy (J)")
plt.xlabel("Total spring length, (m)")
plt.show()
```

```python
plt.plot(x,u_total,label="Total potential energy")
plt.legend()
plt.ylabel("Energy (J)")
plt.xlabel("Total spring length, (m)")
plt.title("This is the potential energy landscape for a\nmass hanging from a spring")
plt.show()
```

## Now, a plot of the potential energy of a swingset

```python
from math import *
m=50 #kg
g=9.8 #N/kg
r=3.0 # chain length, m
max_angle=30

th=[]
u_g=[]
for i in range(31):
    angle=-max_angle+i*max_angle/(30/2)
    gravitational_energy = m*g*(r-r*cos(angle*pi/180))
    th.append(angle)
    u_g.append(gravitational_energy)
plt.plot(th,u_g)
plt.title("The gravitational potential energy of a swingset")
plt.xlabel("angle from down (degrees)")
plt.ylabel("Energy (J)")
plt.show()
```
